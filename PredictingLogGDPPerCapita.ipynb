```python
#set index to be country name
df.set_index('country', inplace=True)

#explore null values
df.isnull().sum()
```

```python
#Pick the features (the columns with less than 100 null values)
features = ['Life Ladder', 'Log GDP per capita', 'Healthy life expectancy at birth', 'Generosity', 'Perceptions of corruption', 'Positive affect', 'Negative affect', 'Confidence in national government']
df = df[features]
```

```python
#drop null values
df.fillna(df.mean(), inplace=True)
df.isnull().sum()
```

```python
#group examples together by country and merge them.
df = df.groupby(['country']).mean().reset_index()
df = df.drop(columns=['country'])
df.head(15)
df.dtypes
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
y = df['Log GDP per capita'] 
X = df.drop(columns = 'Log GDP per capita', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=123)
df.corr()
```

```python
# Create the  LinearRegression model object 
model = LinearRegression()

model.fit(X_train, y_train)

prediction = model.predict(X_test)

#mean squared error
print('\nModel Performance\n\nRMSE =   %.2f'
      % np.sqrt(mean_squared_error(y_test, prediction)))
print(' R^2 =   %.2f'
      % r2_score(y_test, prediction))
```

```python
print('Model Summary:\n')

print('Intercept:')
print('alpha = ' , model.intercept_)

print('\nWeights:')
i = 0
for w in model.coef_:
    print('w_',i+1,'= ', w, ' [ weight of ', features[i],']')
    i += 1
```

```python
#using Stochastic Gradient Descent
from sklearn.linear_model import SGDRegressor

model2 = SGDRegressor(loss='huber', tol=1e-3, learning_rate = 'adaptive')

model2.fit(X_train, y_train)

prediction2 = model2.predict(X_test)

#mean squared error
print('Model Performance\n\nRMSE =   %.2f'
      % np.sqrt(mean_squared_error(y_test, prediction2)))
print(' R^2 =   %.2f'
      % r2_score(y_test, prediction2))
```

```python
#Using Stacking
from sklearn.ensemble import StackingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR

base_models = [
    ('lr1', LinearRegression()),
    ('dt', DecisionTreeRegressor(random_state=42)),
    ('svr', SVR())
]

stacking_model = StackingRegressor(estimators=base_models)
stacking_model.fit(X_train, y_train)

prediction4 = stacking_model.predict(X_test)

#mean squared error
print('Model Performance\n\nRMSE =   %.2f'
      % np.sqrt(mean_squared_error(y_test, prediction4)))
print(' R^2 =   %.2f'
      % r2_score(y_test, prediction4))
```

```python
#Using bagging
from sklearn.ensemble import BaggingRegressor
from sklearn.datasets import make_regression

bagging_model = BaggingRegressor(estimator=LinearRegression(), random_state=42)
bagging_model.fit(X_train, y_train)

prediction3 = bagging_model.predict(X_test)

#mean squared error
print('Model Performance\n\nRMSE =   %.2f'
      % np.sqrt(mean_squared_error(y_test, prediction3)))
# The coefficient of determination: 1 is perfect prediction
print(' R^2 =   %.2f'
      % r2_score(y_test, prediction3))
```

```python
#Performing corss validation
from sklearn.model_selection import cross_val_score, cross_validate
cv_results = cross_val_score(model, X, y, cv=5, scoring='neg_root_mean_squared_error')

print("Cross-Validation RMSE Scores: ", -cv_results)
print("Mean RMSE: ", -cv_results.mean())
```
